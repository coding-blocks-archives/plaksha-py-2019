```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_blobs(n_samples=1000, centers=3, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.classes_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)


in_sample = np.linspace(-4, 12, 10)

for coef, bias in zip(model.coef_, model.intercept_):
    
    out_sample = - (in_sample * coef[0]/coef[1]) - bias/coef[1]
    
    plt.plot(in_sample, out_sample)
```
